```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
excl = pd.read_csv('EXCL.JK.csv', parse_dates = ['Date'], index_col = 'Date')
fren = pd.read_csv('FREN.JK.csv', parse_dates = ['Date'], index_col = 'Date')
isat = pd.read_csv('ISAT.JK.csv', parse_dates = ['Date'], index_col = 'Date')
tlkm = pd.read_csv('TLKM.JK.csv', parse_dates = ['Date'], index_col = 'Date')
```

```python
excl = excl.sort_index()
excl = excl.resample('D').sum()
excl.iloc[:, :5] = excl.iloc[:, :5].replace(0, np.NaN)
excl = excl.fillna(method = 'ffill')
```

```python
fren = fren.sort_index()
fren = fren.resample('D').sum()
fren.iloc[:, :5] = fren.iloc[:, :5].replace(0, np.NaN)
fren = fren.fillna(method = 'ffill')
```

```python
isat = isat.sort_index()
isat = isat.resample('D').sum()
isat.iloc[:, :5] = isat.iloc[:, :5].replace(0, np.NaN)
isat = isat.fillna(method = 'ffill')
```

```python
tlkm = tlkm.sort_index()
tlkm = tlkm.resample('D').sum()
tlkm.iloc[:, :5] = tlkm.iloc[:, :5].replace(0, np.NaN)
tlkm = tlkm.fillna(method = 'ffill')
```

Disediakan 4 buah file .csv yang berisi harga historis saham harian selama 3 bulan terakhir (18/Mar/2019 - 14/Jun/2019) dari 4 perusahaan provider telekomunikasi ternama di Indonesia. 

### **1. SOAL A**
Buatlah sebuah file python yang mengakses data dari ke-4 file .csv tersebut, kemudian tampilkan data plot harga penutupan (close) harian seluruh data yang ada. Output yang diharapkan berupa grafik garis yang menampilkan harga penutupan (close) harian dari ke-4 perusahaan telco selama 3 bulan terakhir:

```python
a = excl['Close']
b = fren['Close']
c = isat['Close']
d = tlkm['Close']

dd = tlkm.index.tolist()

plt.figure(figsize=(11,6), dpi=80, facecolor='w', edgecolor='w')
plt.plot(dd,a,'g', dd,b,'#33ffe9', dd,c,'b', dd,d,'r')
# plt.title('Harga Historis Saham Provider Telco Indonesia', size='20')
plt.suptitle('Harga Historis Saham Provider Telco Indonesia', size='20')
# plt.axis(['2019-03-16', '2019-06-15', 0, 4000])
plt.xticks(rotation=45)
plt.xlabel('Tanggal', size='17') 
plt.ylabel('Rupiah (IDR)', size='17')
plt.grid(True)
plt.legend(['PT XL Axiata Tbk', 'PT Smartfren Telecom Tbk', 'PT Indosat Tbk', 'PT Telekomunikasi Indonesia Tbk'], bbox_to_anchor=(0.,1,1., .102),
          loc='lower left', ncol=4, mode='expand', borderaxespad=0.)
# plt.subplots_adjust(top=1)
plt.show()
```

### **2. SOAL B**
Buatlah sebuah file python yang mengakses data dari ke-4 file .csv tersebut, kemudian tampilkan data plot harga penutupan (close) harian selama bulan April 2019. Output yang diharapkan berupa grafik garis yang menampilkan harga penutupan (close) harian dari ke-4 perusahaan telco selama bulan April 2019:

```python
a2 = a['2019-04']
b2 = b['2019-04']
c2 = c['2019-04']
d2 = d['2019-04']

dd2 = a2.index.tolist()

plt.figure(figsize=(11,6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(dd2,a2,'g', dd2,b2,'#33ffe9', dd2,c2,'b', dd2,d2,'r')
plt.suptitle('Harga Historis Saham Provider Telco Indonesia (April 2019)', size=20)
plt.xticks(rotation=45)
plt.xlabel('Tanggal', size='17') 
plt.ylabel('Rupiah (IDR)', size='17')
plt.grid(True)
plt.legend(['PT XL Axiata Tbk', 'PT Smartfren Telecom Tbk', 'PT Indosat Tbk', 'PT Telekomunikasi Indonesia Tbk'], bbox_to_anchor=(0.,1,1., .102),
          loc='lower left', ncol=4, mode='expand', borderaxespad=0.)
# plt.subplots_adjust(top=1)
plt.show()
```
